```python
start='2009-12-31'
end='2019-12-05'
spy=wb.DataReader('SPY',start=start,end=end,data_source='yahoo')['Adj Close']
```

```python
Symbols=['NFLX','MKTX','ABMD','TDG','AVGO','ALGN','URI',
         'REGN','ULTA','AMZN','EXR','STZ','NVDA','TTWO',
         'ROST','FTNT','MA','CHTR','ORLY','CTAS']
```

```python
df=wb.DataReader(Symbols,start=start,end=end,data_source='yahoo')['Adj Close']
```

```python
df_return=pd.DataFrame(columns=df.columns,index=['return'])
```

```python
for col in df.columns:
    x=df[col].copy().dropna(axis=0)
    df_return.loc['return',col]=(x.iloc[-1]-x.iloc[0])*100/x.iloc[9]
```

```python
df_return.T.plot(kind='bar');plt.ylabel('Cumulative Returns (%)');  
```

```python
spy_return=(spy.iloc[-1]-spy.iloc[0])*100/spy.iloc[0]
spy_return
```

```python
df_return_yearly=pd.DataFrame(columns=df.columns, index=pd.unique(df.index.year))
df_return_yearly=df_return_yearly[df_return_yearly.index>2009]
```

```python
for col in df.columns:
    x=df[col].copy().dropna(axis=0)
    x=x[x.index.year>2009]
    for year in pd.unique(x.index.year):
        x_=x[x.index.year==year].copy()
        df_return_yearly.loc[year,col]=(x_.iloc[-1]-x_.iloc[0])*100/x_.iloc[9]
```

```python
df_return_yearly
```

```python
df_return_yearly[df_return_yearly <= 0].count(axis=0).sort_values().plot(kind='bar');
plt.ylabel('number of years with negative return');
```

```python
spy_return_yearly=pd.DataFrame(columns=['return'],index=pd.unique(spy.index.year))
spy_return_yearly=spy_return_yearly[spy_return_yearly.index>2009]
```

```python
for year in spy_return_yearly.index:
    x_=spy[spy.index.year==year].copy()
    spy_return_yearly.loc[year,'return']=(x_.iloc[-1]-x_.iloc[0])*100/x_.iloc[9]
```

```python
df_return_yearly_normalized=df_return_yearly.div(spy_return_yearly.values, axis='columns')
```

```python
df_return_yearly_normalized[df_return_yearly_normalized >= 1].count(axis=0).sort_values(ascending=False).plot(kind='bar');
plt.ylabel('number of years outperforming SPY');
```

```python
df_return_yearly_normalized[df_return_yearly_normalized >= 1].count(axis=0).sort_values(ascending=False)
```

```python
df_sharpe_spy=pd.DataFrame(columns=['SPY'], index=pd.unique(df.index.year))
df_sharpe_spy=df_sharpe_spy[df_sharpe_spy.index>2009]
x=spy.copy().dropna(axis=0)
x=x[x.index.year>2009]
for year in pd.unique(spy.index.year):
    x_=spy[spy.index.year==year].copy()
    df_sharpe_spy.loc[year,'SPY']=np.nansum(x_.pct_change())*sqrt(252)/np.nanstd(x_)
df_sharpe_spy=df_sharpe_spy.dropna()
```

```python
df_sharpe=pd.DataFrame(columns=df.columns, index=pd.unique(df.index.year))
df_sharpe=df_sharpe[df_sharpe.index>2009]
for col in df.columns:
    x=df[col].copy().dropna(axis=0)
    x=x[x.index.year>2009]
    for year in pd.unique(x.index.year):
        x_=x[x.index.year==year].copy()
        df_sharpe.loc[year,col]=np.nansum(x_.pct_change())*sqrt(252)/np.nanstd(x_)
```

```python
df_sharpe
```

```python
ax=df_sharpe.median(axis=0).sort_values(ascending=False).plot(kind='bar');
ax.axhline(linewidth=4, color='r',y=df_sharpe_spy.median().values[0]);
plt.ylabel('Median Annualized Sharpe Ratio');
```
